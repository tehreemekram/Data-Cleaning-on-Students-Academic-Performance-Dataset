```python
import pandas as pd

# ques1
data1=pd.read_csv("/content/xAPI-Edu-Data.csv")
data1
```

```python
 #Ques2
 print("Handling missing values")
 print(data1.isnull().count())
```

There are no null entries in the dataset so we dont need to handle them.


```python
#Ques 3
print("handling duplicate values")
dupl=data1.duplicated()
data1[dupl]
print(data1.drop_duplicates())
#no duplicated entries in the dataset
```

There are two duplicate row in the dataset,since they do not possess a significant portion of data so they can be removed.

```python
#Ques4
print(data1.head(30))
data1['PlaceofBirth'].replace('KuwaIT','Kuwait',inplace=True)
data1['StageID'].replace('lowerlevel','LowerLevel',inplace=True)
data1['NationalITy'].replace('KW','Kuwait',inplace=True)
data1['NationalITy'].replace('lebanon','Lebanon',inplace=True)
data1['PlaceofBirth'].replace('lebanon','Lebanon',inplace=True)
print(data1.head(30))

data1.rename(columns={'gender':'Gender'},inplace=True)
data1.rename(columns={'NationalITy':'Nationality'},inplace=True)
data1.rename(columns={'VisITedResources':'VisitedResources'},inplace=True)
data1.info()
```

For removing the inconsistencies in data
1.   replaced the I,T in KuwaIT to kuwait because no need of unwanted capitalization.
2.  replaced the lowerlevel to LowerLevel for first word capitalization.
3.  replaced the KW to Kuwait for better understanding.
4.  replaced lebanon to Lebanon in Nationality and placeofbirth for
first word capitalization.
5. renamed the column 'gender':'Gender'for Capitalization.
6. renamed the columns='NationalITy':'Nationality' and 'VisITedResources':'VisitedResources'for removing unwanted capitalization.
